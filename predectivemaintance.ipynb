```python
# delete unnecessary columns
df.drop(['UDI'], axis=1, inplace=True)
```

```python
df.drop(['RNF'], axis=1, inplace=True)
```

```python
df.drop(['OSF'], axis=1, inplace=True)
```

```python
df.drop(['Product ID'], axis=1, inplace=True)
```

```python
df.drop(['HDF','TWF'], axis=1, inplace=True)
```

```python
df
```

```python
df["Type"].unique()
```

```python
Type_dict = {'L': 0, 'M': 1, 'H': 2}
df['Type'] = df['Type'].map(Type_dict)
```

```python
df
```

```python
df['Air temperature [K]'] = df['Air temperature [K]'].round().astype(int)
```

```python
df
```

```python
df['Process temperature [K]'] = df['Process temperature [K]'].round().astype(int)
```

```python
df['Torque [Nm]'] = df['Torque [Nm]'].round().astype(int)
```

```python
df
```

```python
df['Temp_Difference'] = df['Process temperature [K]'] - df['Air temperature [K]']

df['Power_Index'] = df['Rotational speed [rpm]'] * df['Torque [Nm]']

df['Torque_RPM_Ratio'] = df['Torque [Nm]'] / df['Rotational speed [rpm]']
```
